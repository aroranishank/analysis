# Checking correlation between various columns

- based on heatmap below, we can conclude that apparant temperature is proportional to temperature since correlation is 0.99. Thus, apparant temperature increases if temperature increases which in real life is also true.
- When we compare humidity, temperature and apparent temperature, we can see high correlation. Humidity is inversely proportional to apparant temperature and Temperature. That means if temperature/apparent temperature increases, humidity decreases. There is a correlation of -0.56 with temperature and -0.54 with apparent temperature.
- On comparing visibility and temperature, there is high positive correlation of 0.43. that means if temperature increases, visibility increases as well
- On comparing visibility and humidity, there is negative correlation of -0.34. that means if humidity increases, visibility decreases. In real life, it makes sense as well
- There is negative correlation of -0.21 when wind speed and humidity are compared. It makes sense in real life too since if the air is moist, it would not be a little heavy and not easy for air to flow eaily.


```python
mask = np.triu(np.ones_like(df.corr(), dtype=bool))
sns.heatmap(df.corr(), mask=mask,  center=0, linewidths=1, annot=True, fmt=".2f")
```
