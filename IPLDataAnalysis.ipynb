```python
all_season_data = {}

for value in list(df_matches.groupby('season').groups.keys()):
    count=0
    index=0
    total_matches = df_matches[(df_matches['season']==value)]['winner'].count()  

    for index, item in all_winners[all_winners['season']==value].iterrows():
        if(item['toss_winner']==item['winner']):
            count += 1
        index += 1
        percentage = count/total_matches
        all_season_data[value] = percentage


all_season_data
```

```python
import plotly.plotly as py
import plotly.graph_objs as go
import operator

data = []
annotations = []
index=0

for key,value in sorted(all_season_data.items(),key=operator.itemgetter(0), reverse=True):
    data.append( {
      "values": [value, 1-value],
      "labels": [
        "Toss Won Match Won",
        "Toss Lost Match Won"
      ],
      "domain": {"column": index},
      "name": "Match Win %",
      "hoverinfo":"label+percent+name",
      "hole": .4,
      "type": "pie"
    })
    
    
    annotations.append( {
                "font": {
                    "size": 20
                },
                "showarrow": False,
                "text": key,
            })
    if(index%2==0):
        annotations[index]["x"] = 0.20
        annotations[index]["y"] = 0.5
        
    else:
        annotations[index]["x"] = 0.8 
        annotations[index]["y"] = 0.5
        
    if(index==1):
        break;
    index += 1
    

fig = {
  "data": data,
  "layout": {
        "title":"IPL Match Winning % based on toss",
        "grid": {"rows": 1, "columns": 2},
        "annotations": annotations
    }
}
py.iplot(fig, filename='donut')
```

# Chance of winning if batting second 

```python
all_winners[all_winners['season']==2017]
```

```python
df_matches["second_bat_winner"] = False
df_matches["first_bat_winner"] = False
df_matches

```

```python
count_first_bat_winners = 0
count_second_bat_winners = 0
total_matches_in_ipl = df_matches['id'].count()
for index, item in df_matches.iterrows():

    if(item["toss_decision"] == "bat" and item["toss_winner"]!=item["winner"] or (item["toss_decision"] == "field" and item["toss_winner"] == item["winner"])):
        df_matches["second_bat_winner"][index] = True
        count_second_bat_winners += 1
    elif(item["toss_decision"] == "bat" and item["toss_winner"]==item["winner"] or (item["toss_decision"] == "field" and item["toss_winner"] != item["winner"])):
        df_matches["first_bat_winner"][index] = True
        count_first_bat_winners += 1
    else:
        continue    
```

```python
percentage_winning_first_bat = count_first_bat_winners/total_matches_in_ipl
percentage_winning_second_bat = count_second_bat_winners/total_matches_in_ipl


percentage_winning_first_bat
```

```python
percentage_winning_second_bat
```

```python
#plot chart for visualizing winning percentage across various seasons of IPL
import plotly.plotly as py
import plotly.graph_objs as go

fig = {
  "data": [
    {
      "values": [percentage_winning_first_bat, percentage_winning_second_bat],
      "labels": [
        "Team Batting first Winning %",
        "Team Batting second Winning %"
      ],
      "domain": {"column": 0},
      "name": "",
      "hoverinfo":"label+percent+name",
      "hole": .4,
      "type": "pie"
    }],
  "layout": {
        "title":"IPL Winning % based on teams batting first and batting second",
        "grid": {"rows": 1, "columns": 1},
        "annotations": [
            {
                "font": {
                    "size": 20
                },
                "showarrow": False,
                "text": "All IPL seasons"
            }
        ]
    }
}
py.iplot(fig, filename='donut')
```

# Performance of teams on home grounds compared to other grounds

```python
##KOLKATA KNIGHT RIDERS
#total_matches_across_seasons = df_matches[(df_matches.team1 == "Sunrisers Hyderabad") || (df_matches.team2 == "Sunrisers Hyderabad") ]['id'].count()
#df_KKR = pd.DataFrame([],columns=list(df_matches.columns))
df_KKR = pd.DataFrame().reindex_like(df_matches)
#df_KKR.columns = df_matches.columns()
count_KKR_matches=0
for index, item in df_matches.iterrows():
    #print(item)
    if(item['team1']=="Kolkata Knight Riders" or item["team2"] == "Kolkata Knight Riders"):
        count_KKR_matches += 1

count_KKR_matches

```

```python
index1 = 0
#df_KKR = []
for index, item in df_matches.iterrows():
    if(item['team1']=="Kolkata Knight Riders" or item["team2"] == "Kolkata Knight Riders"):
        #df_KKR[index]=item
        #print(bool(item['second_bat_winner']))
        df_KKR['id'][[index1]] = item['id']
        df_KKR['season'][index1] = int(item['season'])
        df_KKR['city'][index1] = item['city']
        df_KKR['date'][index1] = item['date']
        df_KKR['team1'][index1] = item['team1']
        df_KKR['team2'][index1] = item['team2']
        df_KKR['toss_winner'][index1] = item['toss_winner']
        df_KKR['toss_decision'][index1] = item['toss_decision']
        df_KKR['result'][index1] = item['result']
        df_KKR['dl_applied'][index1] = item['dl_applied']
        df_KKR['winner'][index1] = item['winner']
        df_KKR['win_by_runs'][index1] = item['win_by_runs']
        df_KKR['win_by_wickets'][index1] = item['win_by_wickets']
        df_KKR['player_of_match'][index1] = item['player_of_match']
        df_KKR['venue'][index1] = item['venue']
        df_KKR['umpire1'][index1] = item['umpire1']
        df_KKR['umpire2'][index1] = item['umpire2']
        df_KKR['umpire3'][index1] = item['umpire3']
        df_KKR['second_bat_winner'][index1] = bool(item['second_bat_winner'])
        df_KKR['first_bat_winner'][index1] = bool(item['first_bat_winner'])
        index1 += 1
```

```python
#df_KKR.drop(['umpire3'],axis=1,inplace=True)
df_KKR.dropna(inplace=True)
df_KKR
```

```python
matches_won_KKR = df_KKR[df_KKR['winner']=="Kolkata Knight Riders"]['winner'].count()
total_matches = df_KKR['id'].count()
match_win_perc = matches_won_KKR/total_matches
match_win_perc
```

# Heat map with correlation coefficient

```python

import seaborn as sns
import numpy as np
corr=df_matches[['id','season','dl_applied', 'win_by_runs', 'win_by_wickets', 'second_bat_winner', 'first_bat_winner']].corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
# Add the mask to the heatmap
sns.heatmap(corr, mask=mask,  center=0, linewidths=1, annot=True, fmt=".2f")
```
